```python
%matplotlib inline
import numpy
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from sklearn import datasets
import matplotlib.cm as cm
```

```python
### generating two gaussian blobs with the "make_blobs" function form sklearn
data = datasets.make_blobs(n_samples=500,centers=2, random_state=32689)
plt.scatter(data[0][:,0],data[0][:,1],c=data[1])
plt.show()
```

```python
# function that finds the mu and sigma for input matrix x
def gaussian(x):
    mu = numpy.array([numpy.mean(x[:,i]) for i in xrange(x.shape[1])])
    cov = numpy.cov(x)
    return mu,cov
```

```python
#separating the data into the two groups according to class label
data1 = data[0][data[1]==0]
data2 = data[0][data[1]==1]
```

```python
#obtaining the mu and cov matrices
mu1,cov1 = gaussian(data1)
mu2,cov2 = gaussian(data2)
```

```python
### interpolation between data1 and data2.
```

```python
new_data1 = data1-mu1+mu2
to_plot = numpy.vstack((new_data1,data2))
to_ploty = numpy.concatenate((numpy.zeros(250), numpy.ones(250)))
```

```python
plt.scatter(to_plot[:,0],to_plot[:,1],c=to_ploty)
plt.show()
```
